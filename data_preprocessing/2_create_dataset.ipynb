```python
import pandas as pd
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv('/texts.csv', delimiter=",")
#df = pd.read_csv('/words.csv', delimiter=",")
```

```python
df_2 = df[(df['grade'] == 2)]
df_3 = df[(df['grade'] == 3)]
df_4 = df[(df['grade'] == 4)]
df_5 = df[(df['grade'] == 5)]
```

```python
df_2['deficit'] = pd.qcut(df_2['accuracy_pcpm'], q=4, labels=['atencao', 'baixo', 'médio', 'alto'])
df_3['deficit'] = pd.qcut(df_3['accuracy_pcpm'], q=4, labels=['atencao', 'baixo', 'médio', 'alto'])
df_4['deficit'] = pd.qcut(df_4['accuracy_pcpm'], q=4, labels=['atencao', 'baixo', 'médio', 'alto'])
df_5['deficit'] = pd.qcut(df_5['accuracy_pcpm'], q=4, labels=['atencao', 'baixo', 'médio', 'alto'])
```

Stratify

```python
train_2, temp_2 = train_test_split(df_2, test_size=0.4, stratify=df_2['deficit'], random_state=42)
val_2, test_2 = train_test_split(temp_2, test_size=0.5, stratify=temp_2['deficit'], random_state=42)

train_3, temp_3 = train_test_split(df_3, test_size=0.4, stratify=df_3['deficit'], random_state=42)
val_3, test_3 = train_test_split(temp_3, test_size=0.5, stratify=temp_3['deficit'], random_state=42)

train_4, temp_4 = train_test_split(df_4, test_size=0.4, stratify=df_4['deficit'], random_state=42)
val_4, test_4 = train_test_split(temp_4, test_size=0.5, stratify=temp_4['deficit'], random_state=42)

train_5, temp_5 = train_test_split(df_5, test_size=0.4, stratify=df_5['deficit'], random_state=42)
val_5, test_5 = train_test_split(temp_5, test_size=0.5, stratify=temp_5['deficit'], random_state=42)
```

Concat and Save

```python
df_train = pd.concat([train_2, train_3, train_4, train_5], ignore_index=True)
df_train.to_csv(f'train.csv', index=False)
```

```python
df_val = pd.concat([val_2, val_3, val_4, val_5], ignore_index=True)
df_val.to_csv(f'val.csv', index=False)
```

```python
df_test = pd.concat([test_2, test_3, test_4, test_5], ignore_index=True)
df_test.to_csv(f'test.csv', index=False)
```
